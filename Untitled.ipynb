```python
names_file = 'raw_data/babies-first-names-all-names-all-years.csv'
director_file = 'raw_data/director_data.csv'
episodes_file = 'raw_data/episodesTest_0.csv'
movie_file = 'raw_data/movie_data.csv'
tv_show_file = 'raw_data/tvShowTest.csv'
writer_file = 'raw_data/writer_data.csv'
stars_file = 'raw_data/actor_data.csv'

names_df = pd.read_csv(filepath_or_buffer=names_file, index_col=0)
director_df = pd.read_csv(filepath_or_buffer=director_file, index_col=0)
episodes_df = pd.read_csv(filepath_or_buffer=episodes_file, index_col=0)
movie_df = pd.read_csv(filepath_or_buffer=movie_file, index_col=0)
tv_show_df = pd.read_csv(filepath_or_buffer=tv_show_file, index_col=0)
writer_df = pd.read_csv(filepath_or_buffer=writer_file, index_col=0)
stars_df = pd.read_csv(filepath_or_buffer=stars_file, index_col=0)

def check_in_stars(df, stars, cast_type, bad_stars):
    for star in stars:
        if not ((df['name'] == star) & (df['castType'] == cast_type)).any():
            bad_stars.append([star, cast_type])
    return bad_stars

def add_to_cast(df, not_in):
    for cast in not_in:
        if not ((df['name'] == cast[0]) & (df['castType'] == cast[1])).any():
            temp_dict = {'name':cast[0], 'castType':cast[1], 'description':''}
            df = df.append(temp_dict, ignore_index=True)
    return df

def sim_rate(users, content_ids, df):
    good_ratings = ["wow that was such a good movie!",
                    "I am going to recomend this to everyone i know",
                    "this is my new favorite thing to watch",
                    "how did it take so long for someone to have such a revolutional idea",
                    "I wish this app had a bookmark feature so I can save it for later"]
    medium_ratings = ["I mean it's not the worst thing I have ever watched", 
                      "if heaven and hell are real this must be earth, just meh", 
                      "I don't get all the hype, I mean it aight, but like you know", 
                      "I think i need to get HBO",
                      "Well i mean I heard a dude got eye cancer watching something, it wasn't that bad"]
    bad_ratings = ["If i were in a room with Hitler, Stalin and this director and had a gun with 2 bullets id shoot this director twice",
                   "You know when you have no expectations, but yet somehow you are still let down. If not, watch this",
                   "If you just painted your house, you should probably watch that dry instead",
                   "I just wish I could get my time back",
                   "who actually enjoyed this"]
#     my_columns = ['userEmail', 'contentHeaderId', 'review', 'rating']
    ratings = []
    subscriptions = []
    for user in users:
        rated = []
        subs = []
        for i in range(6):
            
            rating = round(random.uniform(0,5),2)
            
            if rating > 3:
                review = random.choice(good_ratings)
            elif rating > 1:
                review = random.choice(medium_ratings)
            else:
                review = random.choice(bad_ratings)
                
            content_id = random.choice(content_ids)
            while content_id in rated:
                content_id = random.choice(content_ids)
            
            serviceName = df.loc[content_id, 'serviceName']
            if serviceName not in subs:
                subs.append(serviceName)
            
            rated.append(content_id)
            add = {'userEmail':user, 'contentHeaderId':content_id, 'review':review, 'rating':rating}
            ratings.append(add)
            
        for sub in subs:
            subscriptions.append({'userEmail':user, 'serviceName':sub})
            
            
    ratings_df = pd.DataFrame(ratings)
    subs_df = pd.DataFrame(subscriptions)
    return ratings_df, subs_df

```
